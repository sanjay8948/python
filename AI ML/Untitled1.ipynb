```python
def get_outlier(data_1):
    new_data = []
    outliers = []
    q1 = data_1.quantile(0.25)
    q3 = data_1.quantile(0.75)
    IQR = q3 - q1
    lv = q1 - 1.5*IQR
    uv = q3 + 1.5*IQR
    for i in data_1:
        if((lv>i) | (uv<i)):
            outliers.append(i)
        else:
            new_data.append(i)
    return outliers,new_data    
```
