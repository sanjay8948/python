```python
#Now creating a function which take dataframe as input and remove outliers from datasets
# This function take two arguments 
# 1- Dataframe
# 2- Index of column in which you want to remove outliers
# Note - According to Question in this function I do not use multivariate approch 
# so we provide this function one more argument that is column index.

def get_outlier(data_1,col_index):
    column = data_1.columns
    q1 = data_1[column[col_index]].quantile(0.25)
    q3 = data_1[column[col_index]].quantile(0.75)
    IQR = q3 - q1
    lv = q1 - 1.5*IQR
    uv = q3 + 1.5*IQR
    outliers = data_1[(data_1[column[col_index]]<lv) | (data_1[column[col_index]]>uv)]
    data_1 = data_1[(data_1[column[col_index]]>lv) &(data_1[column[col_index]]<uv)]
    return outliers,data_1   
```
