## Linear Regression

```python
from sklearn.model_selection import train_test_split
```

```python
#Converting string date to date_time
import datetime as dt
india_case['date'] = pd.to_datetime(india_case['date'])
india_case.head()
```

```python
india_case['date'] = india_case['date'].map(dt.datetime.toordinal)
india_case.head()
```

```python
import numpy as np
```

```python
#getting dependent variable and independent variable
x_data = india_case['date']
x = np.array(x_data).reshape(-1,1)

y_data = india_case['total_cases']
y = np.array(y_data).reshape(-1,1)
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.25)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#y_pred = model.predict(x_test)
x_test
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test,y_pred)
```
