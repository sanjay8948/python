## Example of Poisson Distribution
If on an average, 6 customers arrive every two minutes at a bank during the busy hours of working -
a.) What is the probability that exactly four customers arrives in a given minute?
b.) What is the probability that more than three customers will arrive in a given minute?

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
rate = 6
```

```python
n = np.arange(0,20)
```

```python
poisson = stats.poisson.pmf(n,rate)
```

```python
poisson
```

```python
poisson[3]-poisson[2]
```

```python
1 - (poisson[0]+poisson[1]+poisson[2]+poisson[3])
```

```python
plt.plot(n, poisson, 'o-')
plt.title('Poisson: $\lambda$ = %i' % rate)
plt.xlabel('Number of customers arriving in a minute')
plt.ylabel('Probability of number of customers arriving in a minute')
plt.show()
```
