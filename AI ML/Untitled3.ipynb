```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/h.p/csvfiles/Boston.csv')
data.head()
```

```python
corr_matrix = data.corr().abs()
corr_matrix
```

```python
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(np.bool))
```

```python
print(upper)
```

```python
to_drop = [column for column in upper.columns if any(upper[column]> 0.85)]
print(to_drop)
```

```python
df=data.drop(data[to_drop], axis = 1)
```

```python
df
```

```python
corr = data.corr()
```

```python
columns = np.full((corr.shape[0]),True,dtype = True)
```

```python
columns
```

```python
for i in range(corr.shape[0]):
    for j in range(i+1,corr.shape[0]):
        if corr.iloc[i,j]>=0.85:
            if columns[j]:
                columns[j]=False
selected_columns = data.columns[columns]
```

```python
selected_columns
```

```python
data = data[selected_columns]
```
