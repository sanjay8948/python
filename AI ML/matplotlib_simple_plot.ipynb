```python
import numpy as np
from numpy import pi
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,2,100)
```

```python
plt.plot(x,label="Linear")
plt.plot(x**2,label="Quadratic")
plt.plot(x**3,label="Cubic")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
# plotting sin wave

x=np.linspace(0,2*np.pi,40)
y=np.sin(x)
```

```python
plt.plot(y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Plotting Sin(x)")
plt.show()
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0,5,0.1)
t2=np.arange(0,5,0.02)
plt.figure()
plt.subplot(3,1,1)     # (nrow,ncolumn,index)
plt.plot(t1,f(t1),"bo",t2,f(t2),"red")
plt.subplot(3,1,3)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```

```python
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)
y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)
```

```python
plt.subplot(2,1,1)
plt.plot(x1,y1,'og-')
plt.xlabel("x1")
plt.ylabel("y1")
plt.title('Subplot-1')

plt.subplot(2,1,2)
plt.plot(x2,y2,'o-')
plt.xlabel('x2')
plt.ylabel('y2')
plt.title('Subplot-2')
plt.show()
```

```python
#Bar graph

x=[1,2,3,4,5]
y=[5,8,12,40,30]
tick_label=['one','two','three','four','five']
```

```python
plt.bar(x,y,tick_label=tick_label,width=0.8,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Plot')
plt.show()
```

```python
#Histogram

ages=[2,50,70,40,30,45,50,45,43,40,44,60,7,13,51,18,20,77,32,21,20,40]
range=(0,100)
bins=10
```

```python
plt.hist(ages,bins,range,histtype='bar',rwidth=0.8)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Histogram Plot')
plt.show()
```

```python
#Scatter plot

x=np.random.random(50)
y=np.random.random(50)
```

```python
plt.scatter(x,y,label='stars',color='violet',marker='*',s=100)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
#piechart

activities=['eat','sleep','work','play']
slices=[3,7,6,9]
colors=['r','b','m','b']
```

```python
plt.pie(slices,labels=activities,colors=colors,shadow=True,explode=(0.2,0,0,0),autopct='%0.2f%%')
plt.legend()
plt.show()
```
