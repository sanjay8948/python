```python
import numpy as np
```

```python
x = np.random.rand(10,4)
x
```

```python
mean = np.array([np.mean(x, axis = 0)])
mean
```

```python
std = np.array([np.std(x , axis = 0)])
std
```

```python
mean_repeated = np.repeat(mean, [10], axis = 0)
mean_repeated
```

```python
std_repeated = np.repeat(std, [10], axis = 0)
std_repeated
```

```python
X = x - mean_repeated
X
```

```python
X = x / (std_repeated + 0.0000001)
X
```
