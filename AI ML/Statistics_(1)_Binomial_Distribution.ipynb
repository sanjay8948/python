## Example for Binomial Distribution
A bank issues credit cards to customers under the scheme of MasterCard. Based on the past data, the bank has found out that 60% of all accounts pay on time following the bill. If a sample of 7 accounts is selected at random from the current database, construct the Binomial Probability Distribution of accounts paying on time.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
n = 7
```

```python
p = 0.6
```

```python
k = np.arange(0,8)
k
```

```python
binomial = stats.binom.pmf(k,n,p)
```

```python
binomial
```

```python
plt.plot(k,binomial,'o-')
plt.title('Binomial: n = %i, p = %.2f' %(n,p), fontsize = 15)
plt.xlabel('Number of successes')
plt.ylabel('Probability of successes')
plt.show()
```
